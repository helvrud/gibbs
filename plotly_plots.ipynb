```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px

data_path = "data/gel_all_data.pkl"

df = pd.read_pickle(data_path).dropna()
```

```python
df
```

```python
plot_data = df.loc[df['delta_P_Pa_mean'] < 6e5]
plot_data['sample_size'] = plot_data['anion_salt'].apply(len)
plot_data = plot_data.loc[plot_data['sample_size'] >30]
plot_data['c_s_reservoir_mol'] = plot_data['c_s_reservoir_mol'].astype('category')
plot_data['box_l'] = plot_data['V0']**(1/3)
px.scatter(
    plot_data, 
    x = 'gel_density', log_x = 'True', 
    y = 'c_s_mol_mean',
    error_y = 'c_s_mol_err',
    color = 'c_s_reservoir_mol',
    hover_data=['n_pairs', 'v', 'V0', 'box_l', 'sample_size', 'delta_P_Pa_mean'],
    )
```

```python
plot_data = df.loc[df['delta_P_Pa_mean'] < 6e5]
plot_data['sample_size'] = plot_data['anion_salt'].apply(len)
plot_data = plot_data.loc[plot_data['sample_size'] >30]
plot_data['c_s_reservoir_mol'] = plot_data['c_s_reservoir_mol'].astype('category')
plot_data['box_l'] = plot_data['V0']**(1/3)
px.scatter(
    plot_data, 
    x = 'gel_density', log_x = 'True', 
    y = 'delta_P_Pa_mean',
    error_y = 'delta_P_Pa_err',
    color = 'c_s_reservoir_mol',
    hover_data=['n_pairs', 'v', 'V0', 'box_l', 'sample_size']
    )
```

```python
from plotly.subplots import make_subplots
import plotly
cols = plotly.colors.DEFAULT_PLOTLY_COLORS
fig = make_subplots(rows=2, cols=1, shared_xaxes=True, )
for color, (idx, group) in zip(cols, plot_data.groupby(by = 'c_s_reservoir_mol')):
    trace_P = go.Scatter(
        x = group.gel_density, 
        #log_x = 'True', 
        y = group.delta_P_Pa_mean*1e-5,
        name = idx,
        error_y = dict(type ='data', array = group.delta_P_Pa_err*1e-5),
        #color = idx,
        text=group.delta_P_Pa_mean,
        hovertemplate='<br>x:%{x}<br>y:%{y}<br>delta_P:%{text} Pa',
        mode = 'markers',
        marker = dict(color = color),
    )
    trace_c_s = go.Scatter(
        x = group.gel_density, 
        #log_x = 'True', 
        y = group.c_s_mol_mean,
        name=idx,
        error_y = dict(type ='data', array = group.c_s_mol_err),
        #color = idx,
        #hover_data=['n_pairs', 'v', 'V0', 'box_l', 'sample_size'],
        text=group.c_s_mol_mean,
        hovertemplate='<br>x:%{x}<br>y:%{y}<br>c_s:%{text} mol/l',
        mode = 'markers',
        marker = dict(color = color),
    )
    fig.add_trace(
        trace_c_s,
        row = 1, col = 1,
    )
    fig.add_trace(
        trace_P,
        row = 2, col = 1
    )

fig.update_xaxes(
    showspikes = True, spikemode='across',
    #title_text="x-axis in logarithmic scale", 
    type="log", 
    row=1, col=1
    )
fig.update_xaxes(
    showspikes = True, spikemode='across',
    title_text="gel density", 
    type="log", 
    row=2, col=1
    )
fig.update_yaxes(
    showspikes=True,
    title_text="c_s", 
    type="log", 
    row=1, col=1)
fig.update_yaxes(
    showspikes=True,
    title_text="delta_P", 
    #type="log", 
    row=2, col=1)
#fig.update_layout({ax:{"showspikes":True} for ax in fig.to_dict()["layout"] if ax[0:3]=="xax"})
fig.update_traces(xaxis="x1")
fig.update_layout(
    #hovermode="x unified", 
    height = 900, 
    width = 900,)
fig.write_html('track_c_s_and_pressure.html')
```
