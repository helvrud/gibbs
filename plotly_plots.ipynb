```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px

data_path = "data/gel_all_data.pkl"

df = pd.read_pickle(data_path).dropna()
```

```python
df
```

```python
#df['c_s_reservoir_mol'] = df['c_s_reservoir_mol'].astype(str)
plot_data = df.loc[df['delta_P_Pa_mean'] < 5e5]
px.scatter_3d(
    plot_data, 
    x = 'gel_density', log_x = 'True', 
    y = 'c_s_reservoir_mol', log_y = 'True',
    z = 'delta_P_Pa_mean', log_z = 'True', 
    color = 'c_s_reservoir_mol',
    hover_data=['n_pairs']
    )
```

```python
plot_data = df.loc[df['delta_P_Pa_mean'] < 6e5]
plot_data['box_l'] = plot_data['V0']**(1/3)
px.scatter(
    plot_data, 
    x = 'gel_density', log_x = 'True', 
    y = 'delta_P_Pa_mean',
    color = 'c_s_reservoir_mol',
    hover_data=['n_pairs', 'v', 'V0', 'box_l']
    )
```

```python
plot_data = df.loc[df['delta_P_Pa_mean'] < 5e5]
px.scatter_3d(
    plot_data, 
    x = 'v', log_x = 'True', 
    y = 'c_s_reservoir_mol', log_y = 'True',
    z = 'delta_P_Pa_mean', log_z = 'True', 
    color = 'c_s_reservoir_mol',
    hover_data=['n_pairs']
    )
```

```python
df_sorted = df.loc[df['delta_P_Pa_mean'] < 5e5].sort_values(by = 'v')


df_sorted.groupby(by = 'c_s_reservoir_mol').agg({'v' : min})
```
